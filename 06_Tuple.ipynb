## Python Tuple
       A Python tuple is an ordered, immutable sequence type used to store collections of items.
    
    -> Key Characteristics of Tuples
    
    1. Ordered: Items maintain a defined sequence, which does not change.
    
    2. Immutable: Once a tuple is created, you cannot change, add, or remove items. 
       This is the main difference from a list.
    
    3. Heterogeneous: Like lists, tuples can hold items of different data types (e.g., integers, strings, and lists).
    
    4. Efficiency: Because tuples are immutable, Python can optimize their storage and access, 
       often making them slightly faster than lists for iteration.
    
    5. duplicates are allowed.

```python
# create a tuple
var = ()
print(type(var)) 
```

```python
# create a tuple with items
var = (1,2,3,4,5)
print(var) 
```

```python
# create tuple with different types of items
var = (1,2,3,4,5,True,"Like")
print(var) 
```

```python

# create tuple using without parantesis
var = 1,2,3,4,True,"Like"
print(var) 
```

```python
# check data type of items
var = 1,"Like"
print(type(var)) 
```

```python
# find the length of tuple
var = 1,2,3,4,True,"Like"
print(len(var)) 
```

```python
# Find Index value of Tuple
var = 1,2,3,4,True,"Like"
l = var.index(4)
print(l) 
```

```python
# max() : get maximum value 
tup = (1,3,4,6,5)
tup1 = ("Zebra","Dog","Elephant")
tup2 = ("Apple","Orange","Mango")
print(max(tup))
print(max(tup1))
print(max(tup2)) 
```

```python
# min() :: opposite to max()
tup = (1,3,4,6,5)
tup1 = ("Zebra","Dog","Elephant")
tup2 = ("Apple","Orange","Mango")
print(min(tup))
print(min(tup1))
print(min(tup2))
```

```python
# len() :: get length of tuple  
tup = (1,3,4,6,5)
tup1 = ("Zebra","Dog","Elephant")
tup2 = ("Apple","Orange","Mango")
print(len(tup))
print(len(tup1))
print(len(tup2))
```

```python
# tuple() :: convert list into tuple
list = [1,3,5,7]
print("List Item ::",list)
tup = tuple(list)
print("Tuple Item ::",tup) 
```
