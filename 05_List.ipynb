## Python List 

       A Python list is a built-in, versatile sequence data type used to store a collection of items. Lists are one of 
       the most fundamental and frequently used structures in Python.
    
        -> Key Characteristics of Lists
        
        1. Ordered: The items are stored in a specific sequence, and the order will not change. 
           You can access items using their index (position).
        
        2. Mutable: Unlike strings and tuples, lists can be changed after they are created. 
           You can add, remove, and modify elements.
        
        3. Heterogeneous: Lists can store items of different data types simultaneously 
           (e.g., integers, strings, floats, and even other lists).
        
        4. Dynamic: Lists can grow or shrink in size as elements are added or removed.
        
        5. duplicates are allowed.
        
        -> Basic Operation On List
        # Creating a List
        # Accessing List Elements
        # Adding Elements to List
        # Updating Elements into List
        # Removing Elements from the List
        # Iterating Over Lists

```python
List = [1,2 ,4 ,5,"Bob","Alice",True,7.8,5.7]
print(List)

List[4] = "Yug"   #For Updating the List
print(List)

List1 = [8,"rubi",8]
List = List + List1   # Adding Two List
print(List)
```

```python
#  Accessing Element 
print(List[2])
print(List[-1])
```

```python
# #Slicing
print(List[1:])
print(List[:7])
print(List[-6:-3])
```

```python
# Adding Element In a List
List.append("Gaming")      #Add single Element at End.
List.insert(0,"Line")      #Insert Element At specific index.
List.extend([2,4,5,6,8])   # Adds multiple elements to the end of the list.
print(List)
```

```python
#Removing Elements from the List
List.remove("Gaming")  # Remove Value That mentioned .
List.pop(0)            # Remove Value of Index that mentioned.
print(List)
```

```python
# Iterating List using 'for' loop:
for x in List :
    if x=="rubi" :
        break 

    print(x)
```

```python
# Sorting In list   
List2 = [1,5,7,3,9,1,5,10]
List2.sort()
print(List2)
List_2 = List2.copy()
List_2.sort(reverse=True)
print(List_2)
```

```python
# count() For Getting count of an Element in a List
print(List2.count(10))
 
# index() :: For getting an Index of an Element  
print(List2.index(10))

# List compression 
List4 = ["Alice","Bob","Adam","Ani","John","Garry","ava"]
a = [word for word in List4 if word.startswith('A')]
print(a)
```

```python
# List Unpacking 
List5 = [1,4,6]
z,u,v= List5
print(z)
print(u)
print(v)
```

```python
# max() function :: Get maximum value from the list 
# if there is string max value get by ascii 
list_1=["Apple","Orange","Mango"]
list1=["Zebra","Dog","Elephant"]
list2 = [1,6,3,8,3]
print(max(list_1))
print(max(list1))
print(max(list2))
```

```python
# min() function :: Opposite to max() 
list_1=["Apple","Orange","Mango"]
list1=["Zebra","Dog","Elephant"]
list2 = [1,6,3,8,3]
print(min(list2))
print(min(list1))
print(min(list2))
```

```python
# list () function :: convert list into tuple
tup = (1,4,2,7,"kop")
print("Tuple Item ::",tup)
tup = list(tup)
print("List Item ::",tup) 
```
