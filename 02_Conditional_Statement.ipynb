## Conditional Statement  
     - Python's conditional statements allow you to execute different blocks of code based on whether specific conditions 
       are True or False. The primary conditional statements are the if, elif (else if), and else keywords.
    
     1. IF Statement
        The most basic form. The block of code indented below the if is executed only if the condition is True.
    
     2. If - Else Statement 
        This structure provides an alternative block of code to execute if the if condition is False.
    
     3. If - Elif -Else Statement 
        This is used to check multiple conditions sequentially.
     -  The first if condition is checked.
     -  If it's False, the first elif condition is checked.
     -  This continues down the line of elif statements.
     -  If all preceding conditions are False, the else block (if present) is executed.

#### If - Else Statement

```python
# Define credentials and login status
username = "admin"
password = "1234"
is_logged_in = True

# Check if username and password are correct, and if the user is logged in
if (username == "admin" and password == "1234") and is_logged_in:
    print("Access granted.")  # All conditions met
else:
    print("Access denied.")   # One or more
```

#### If - Elif - Else Statement 

```python
# Define financial information
income = 55000           # Annual income of the applicant
credit_score = 700       # Credit score of the applicant
debt = 10000             # Existing debt of the applicant

# Check if the applicant meets the criteria for loan approval
if income >= 50000 and credit_score >= 700 and debt < 15000:
    # If income is 50,000 or more, credit score is 700 or more, and debt is under 15,000
    print("Loan approved.")
    
# If the applicant doesn't meet the top criteria, check if they qualify for review
elif income >= 40000 and credit_score >= 650:
    # If income is at least 40,000 and credit score is at least 650
    print("Loan under review.")
    
# If none of the conditions are met, deny the loan
else:
    print("Loan denied.")
```

#### Ladder If Statement 

```python
 # Taking three numbers as input from the user
a = int(input("Enter the Number:: "))
b = int(input("Enter the Number:: "))
c = int(input("Enter the Number:: "))

# Check if 'a' is greater than both 'b' and 'c'
if a > b:
    if a > c:
        print("A is Larger")

# Check if 'b' is greater than both 'a' and 'c'
if b > a:
    if b > c:
        print("B is Larger ")

# Check if 'c' is greater than both 'a' and 'b'
if c > a:
    if c > b:
        print("C is Larger ")
```

#### Discount Calculator 

```python
# Input the purchase price of the TV
TV_par = float(input("Purchase price :: "))

# Input the selling price of the TV
TV_sa = float(input("Selling price :: "))

# Input the total additional charges (like shipping, service, etc.)
Add_price = float(input("Enter total additional charges :: "))

# Calculate the profit before subtracting additional charges
profit = TV_sa - TV_par

# Subtract additional charges from the profit to get net profit
Total_profit = profit - Add_price

# Display the net profit
print("Our net profit is :: ", Total_profit)

# Apply discount based on the amount of net profit
if Total_profit > 1000:
    # If net profit is greater than 1000, give 5% discount
    D1 = Total_profit * 5 / 100
    print("Discount is :: ", D1)

elif Total_profit > 500:
    # If net profit is greater than 500, give 2.5% discount
    D2 = Total_profit * 2.5 / 100
    print("Discount is :: ", D2)

elif Total_profit > 250:
    # If net profit is greater than 250, give 1% discount
    D3 = Total_profit * 1 / 100
    print("Discount is :: ", D3)

else:
    # If profit is 250 or less, no discount is given
    print("Sorry ....")
    print("You don't get any discount")
```

#### Bank Interface

```python
# Print ATM welcome header.
print("--------------------------------------------------")
print("              BANKING SYSTEM - ATM")
print("--------------------------------------------------")

# Set the correct PIN and initial current balance.
PIN = 1918
cb = 10000  # Current Balance.

# Ask the user to enter their PIN.
E_PIN = int(input("Enter the PIN :: "))

# Check if the entered PIN matches the correct PIN.
if E_PIN == PIN:
    # Display ATM options menu.
    print("--------------------------------------------------")
    print("1. Check Balance")
    print("2. Withdraw Amount")
    print("3. Deposit Amount")
    print("4. Account Information")
    print("5. Cancel")
    print("--------------------------------------------------")

    # Take the user's choice as input.
    cho = int(input("Enter your Choice :: "))

    # If user chooses to check balance.
    if cho == 1:
        print("--------------------------------------------------")
        print("Check Balance")
        print("--------------------------------------------------")
        print("Current Balance :: ₹", cb)
        print("--------------------------------------------------")

    # If user chooses to withdraw money.
    elif cho == 2:
        print("--------------------------------------------------")
        print("Withdraw Amount")
        print("--------------------------------------------------")
        print("Current Balance :: ₹", cb)
        aw = float(input("Enter the Withdrawal Amount :: ₹"))
        
        # Check if there is sufficient balance.
        if aw <= cb:
            cb = cb - aw  # Deduct withdrawal amount from balance.
            print("Amount of ₹", aw, "successfully withdrawn.")
            print("Remaining Balance :: ₹", cb)
        else:
            print("Insufficient Balance!")  # Not enough funds.
        print("--------------------------------------------------")

    # If user chooses to deposit money.
    elif cho == 3:
        print("--------------------------------------------------")
        print("Deposit Amount")
        print("--------------------------------------------------")
        ad = float(input("Enter the amount to Deposit :: ₹"))
        cb = cb + ad  # Add deposit amount to balance.
        print("Amount of ₹", ad, "successfully deposited.")
        print("Updated Balance :: ₹", cb)
        print("--------------------------------------------------")

    # If user wants to see account info.
    elif cho == 4:
        print("--------------------------------------------------")
        print("Account Information")
        print("--------------------------------------------------")
        print("Account Holder Name  :: Satvik Landge")
        print("Account Number       :: xxxxxx00xx")
        print("Bank Name            :: Bank of India")
        print("IFSC Code            :: fsdxxxxxxxx")
        print("--------------------------------------------------")

    # If user chooses to cancel/exit.
    elif cho == 5:
        print("Thank you for using the ATM. Goodbye!")
        exit()  # Exit the program.

    # If user enters an invalid choice.
    else:
        print("Invalid Choice. Try again.")

# If the entered PIN is incorrect.
else:
    print("Wrong PIN entered. Access Denied.")
```
