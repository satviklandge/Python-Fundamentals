## Python Dictionary
    Dictionary is a data structure in which we represent a group of object as key-value pair.
       it's designed to retrieve values based on a unique key, rather than by an index (like lists or tuples).
    
    -> Key Characteristics of Dictionaries
    
    1. Key-Value Pairs: Every item in a dictionary consists of a key and its corresponding value.
    
    2. Unordered: In Python versions 3.7 and later, dictionaries maintain the insertion order, meaning the order 
       in which items are added is preserved. However, historically and fundamentally, they are not ordered by index.
    
    3. Mutable: You can change, add, or remove items after the dictionary is created.
    
    4. Keys Must be Unique and Immutable:
       - Each key must be unique. If you assign a value to an existing key, the old value is overwritten.
       - Keys must be immutable data types (e.g., strings, numbers, or tuples). Lists and other dictionaries cannot be used as keys.
    
    5. Values: Values can be any data type and can be duplicates.

```python
# Empty Dictionary
dic = {}
print(type(dic))
```

```python
dic = {"Name" : "laii","marks" : 34,"Roll_No" : 34}
print(type(dic))
print(dic)
```

```python
# capture the key value using list "[]" index
dic = {"Name" : "laii","marks" : 34,"Roll_No": 35}
print(dic["marks"])
    # OR 
dic = {"Name" : "laii","marks" : 34,"Roll_No": 35}
print(dic.get("Roll_No"))
```

```python
# IF there no key in dictionary set the default message
print(dic.get("Grade","Not Available"))
```

```python
# pop() method - remove key-pair value
dic = {"Name" : "laii","marks" : 34,"Roll_No": 35}
dic.pop("Roll_No")
print(dic)
```

```python
# key() method - All keys to show
dic = {"Name" : "laii","marks" : 34,"Roll_No": 35}
print("Keys :: ",dic.keys())
```

```python
# values() method - All dictionaty values to show
dic = {"Name" : "laii","marks" : 34,"Roll_No": 35}
print("Values  :: ",dic.values())
```

```python
dic = {"Name" : "laii","marks" : 34,"Roll_No": 35}
for key,value in dic.items() :
 print(key,value,sep=" - ")
```

```python
dic = {"Name" : "laii","marks" : 34,"Roll_No": 35}
dic["marks"] = 100
for key,value in dic.items() :
     print(key,value,sep=" - ")
```
