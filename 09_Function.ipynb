## Python Function 
     A function in Python is a block of organized, reusable code that is used to perform a single, related action. 
       Functions provide better modularity for your application and a high degree of code reusing
       Functions are defined using the def keyword, followed by a function name, parentheses (), and a colon :.
    
    -> Key Concept 
    
    1. Parameters and Arguments
    Parameters are the variables listed inside the parentheses in the function definition (e.g., name in the example above).
    Arguments are the actual values passed to the function when it is called (e.g., "Alice").
    
    2. The return Statement
    The return statement is used to exit a function and hand back a value (or values) to the caller.
    If a function doesn't explicitly use a return statement, it automatically returns the special value None.


```python
# userdefine function() ::
a = 2
b = 4
def cal() :
    print("Addition",a+b)
    print("Subtraction",a-b)
    print("Multiplication",a*b)
    print("Division",a/b)

cal()
```

```python
# With Parameter
def cal(a,b):
    print("Addition",a+b)
    print("Subtraction",a-b)
    print("Multiplication",a*b)
    print("Division",a/b)

cal(4,8)
```

```python
#function() for simple interest
def simple_interest(p,r,t):
    ST = (p*r*t)/100
    print("Simple Interest is ::",ST)

p = float(input("Enter the Principle Amount ::"))
r = float(input("Enter Rate of Interest ::"))
t = int(input("Enter Time of Saving ::"))   
    
simple_interest(p,r,t) 
```

```python
# function() for finding largest Number 
def Large_NO(a,b,c) :
    if a>b and a>c :
        print("A is Large")

    elif b>a and b>c :
        print("B is Large") 

    else:
        print("C is large")       
    
Large_NO(100,70,50)
```

#### Pre - Define Maths Functions

```python
# sin() :: It is used to get sine of any number in radians.
import math
x = math.sin(4)
print(x) 
```

```python
# cos () ::: It is used to get cosine of any number in radians.
x = math.cos(2)
print(x) 
```

```python
# tan() :: It is used to get tangent of any number in radians.
x = math.tan(4)
print(x) 
```

```python
# sqrt() :: square root of any number
x = math.sqrt(100)
print(x) 
 
```

```python
# log10() :: base 10 logarithm of any number 
x = math.log10(30)
print(x) 
```

```python
 # pow() :: It is used to get the power of any number.
# if pow(a,b) then it returns a raised to the power b.
x = math.pow(4,2)
print(x)
```

```python
#factorial() :: factorial of any number 
x = math.factorial(7)
print("Factorial of 7 is ::",x) 
```

```python
# exp() :: exponential of number
x=math.exp(0)
print("exp(0):",x) 
```

```python
#floor() :: It returns the largest integer less than or equal to given number.
x = math.floor(4.9)
print(x) 
```

```python
# ceil() :: It returns the smallest integer greater than or equal to given number.
x = math.ceil(6.8)
print(x) 
```
