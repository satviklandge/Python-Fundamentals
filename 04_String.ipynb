## String
       A Python string is a sequence of characters that is used to store and represent text-based information.
    
    -  Key Characteristics
    
    1. Sequence Type: Strings are sequences, meaning they maintain a specific order of characters. 
       Each character in the string can be accessed using its index (position).
    
    2. Immutability: Once a string is created, it cannot be changed. If you perform an operation that seems 
       to "modify" a string (like concatenation or replacement), Python actually creates a brand new string in memory.
    
    3. Creation: Strings are created by enclosing text in single quotes ('...'), double quotes ("..."), 
       or triple quotes ('''...''' or """...""").

#### String Method

```python
# capitalize()
#it converts the first letter of the string into uppercase
str = "homeland"
print("Before",str)
print("After",str.capitalize())  
```

```python
#upper()
# it converts the All letters of the string into uppercase
str = "Homeland"
print("Before",str)
print("After",str.upper()) 
```

```python
#casefold()  :: it converts string into lowecase
str = "HOMeland"
print("Before",str)
print("After",str.casefold()) 
```

```python
# ceter()
# it is used to align the string to the center
# it has two parameter width and fillchar in which fillchar optional.
str = "homeland"
print("Original ::",str)
print("Align ::",str.center(30))
print(str.center(10,"*")) 
```

```python
#endswith() :: it return boolean value(True/False)
# it the given string ends with specified string returns true ontherwise false.
str = "Its my homeland"
print(str)
print(str.endswith("homeland")) 
```

```python
#startswith() :: it return boolean value(True/False)
# it the given string starts with specified string returns true ontherwise false.
str = "Its my homeland"
print(str)
print(str.startswith("It")) 
```

```python
# find()
# it is used to search the specific string in a string.
# if the specificed string is found then it returns the position of where it is found.
str = "Its my homeland"
print(str)
print(str.find("my"))
print(str.find("my",2))  
```

```python
#index()
str = "Its my homeland"
print("Value:",str)
print(str.index("my"))
print(str.index("my",3)) 
```

```python
# format() method :: 
str1 = (2+34+45+3)
str2 = (90-43+3)
str3 = (34+12)
print("{} and {} both are best friend ".format("Bhavesh","Abhishek"))
print("{a} and {b} good luck".format(a="Virat",b="Rohit"))
print("Value is {} and {} and {}".format(str1,str2,str3)) 
```

```python
# isalnum()
# isalpha()
# isdecimal()
# isdigit()
str1 = "game123"
str2 = "game@1234"
str3 = "1234"
str4 = '1233'
print("Is String 1 Alphanumeric ::",str1.isalnum())
print("Is String 2 Alphanumeric ::",str2.isalnum()) 
print("Is String 1 Alphabet ::",str1.isalpha())
print("Is String 2 Alphabet ::",str2.isalpha()) 
print("Is String 1 Decimal ::",str1.isdecimal())
print("Is String 3 Decimal ::",str3.isdecimal())
print("Is String 1 a Digit ::",str1.isdigit())
print("Is String 4 a Digit ::",str4.isdigit()) 
```

```python
# The .isidentifier() method checks if a string is a
# valid Python identifier â€” meaning a name you can use for variables, functions, classes, etc.
str1 = "game123"
str2 = "game@1234"
str3 = "1234"
print(str1.isidentifier())
print(str2.isidentifier())
print(str3.isidentifier()) 
```

```python
# isupper() ::Returns True if all cased characters
# in the string are uppercase, and there is at least one cased character.
print("HELLO".isupper())     # True
print("Hello".isupper())     # False (not all uppercase)
print("123!".isupper())      # False (no letters at all)
print("HELLO123".isupper())  # True (letters are uppercase; digits are ignored)
 
```

```python
# islower() ::Returns True if all cased characters in the string
# are lowercase, and there is at least one cased character.
print("hello".islower())     # True
print("Hello".islower())     # False (H is uppercase)
print("123!".islower())      # False (no lowercase letters)
print("hello123".islower())  # True 
```

```python
# isspace() :: Returns True if all characters in the string 
# are whitespace (spaces, tabs, newlines, etc.), and at least one character is present.
print(" ".isspace())         # True (single space)
print("\t\n".isspace())      # True (tab and newline)
print("  \t  ".isspace())    # True
print(" a ".isspace())       # False (contains a non-space character)
print("".isspace())          # False (empty string)
 
```

```python
# replace() :: The str.replace() method is used to replace parts of a string with something else.
text = "I like apples"
print("Orignal Text ::",text)
print("Replaced Text ::",text.replace("apples", "bananas"))
 
text1 = "one one one "
print("Orignal Text ::",text1)
print("Replaced Text ::",text1.replace("one", "five",2)) 

text3 = "hell0 Impact"
print("Orignal Text",text3)
print("Replaced Text",text3.replace("h","J")) 
```

```python
# len() :: length of string
text = "hello"
print(len(text))    
  
 
# swapcse() :: swap the uppercase with lowercase and viceversa
str = "ApplE"
print(str.swapcase()) 
```

```python
# strip() :: Removes leading and trailing whitespace (spaces, tabs, newlines) by default.
text = "  hello world  "
text1 = "___hello___"
print(text.strip())   
print(text1.strip("_")) 
```

```python
# lstrip() :: Removes leading (left-side) whitespace or specified characters.
text = "  hello  "
text1 = "---test---"
print(text.lstrip())  
print(text1.lstrip("-"))   
```

```python
# rstrip() :: Removes trailing (right-side) whitespace or specified characters.
text = "  hello  "
text1 = "---test---"
print(text.rstrip())    
print(text1.rstrip("-"))
```

```python
# split() :: 
text = "hello world"
print(text.split()) 
```

```python
# Accessing tring element by Index     
print(str1[1])     
print(str1[:3])   
print(str1[-4:]) 
```

```python
# not in :: membership function return true if a particular substring is not present in string
print('sre' not in str1)
print('Da' in str1) 
```

```python
# string formating using % operator
integer = 1
float = 1.2
string = "Pogo "
print ("My value is %d \nMy value is %f \nMy value is %s"%(integer,float,string)) 
```
