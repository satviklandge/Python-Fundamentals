### Loops
     - A loop in programming is a control flow statement that allows a sequence of instructions (a block of code) 
       to be repeatedly executed based on a specific condition or for a predetermined number of iterations.
    
     - Python has two primary types of loops for executing a block of code repeatedly: the for loop and the while loop.
    
    1. for loop
       A for loop is used for iterating over a sequence (such as a list, tuple, dictionary, set, or string) or other iterable objects. 
       It executes the code block once for each item in the sequence.
    
    2. While loop
       A while loop executes a set of statements as long as a specified condition is true.

#### 1. For Loop

```python
#Take input from the user and convert it to an integer
N = int(input("Enter the Number :: "))

# Print a heading for the multiplication table
print(N, "Table")

# Initialize a variable 'a' to store each result (not strictly necessary but kept from original)
a = 1

# Loop from 1 to 10 to generate the multiplication table
for x in range(1, 11):
    a = N * x                  # Multiply the number with the current loop value
    print(a)                   # Print the result
```

#### 2. While Loop

```python
# Using while loop to iterate through a list
fruits = ["apple", "banana", "cherry"]
i = 0
while i < len(fruits):  # Check The Condition
    print(fruits[i])
    i += 1              # Increment Current Value
```

#### 3. Infinite While Loop

```python
# Example of an infinite loop with a break condition
while True:
    user_input = input("Type 'exit' to stop: ")
    if user_input == "exit":
        break  # Exit the loop
    print("You typed:", user_input)
```
