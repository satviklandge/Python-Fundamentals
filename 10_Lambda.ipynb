## lambda Function 
    A Lambda Function in Python is a small, anonymous function defined with the lambda keyword. 
       It's an inline function that can take any number of arguments but can only have one expression.
    
    -> Key Characteristics
    
    1. Anonymous: It doesn't have a formal name defined using def.
    
    2. Single Expression: It must contain only one executable expression. The result of this expression is automatically returned.
    
    3. Concise: They are syntactically restricted and are generally used for short, throw-away functions, often as arguments 
       to higher-order functions like map(), filter(), or sorted().
    
    Syntax :
    lambda arguments: expression

```python
# we gave an example of a lambda function that adds 4 to the
# input number is shown below 
add = lambda D : D + 7
print(add(4)) 
```

```python
# Now we gave an example of a lambda function that multiply 
# 2 numbers and return one result
multi = lambda x ,y : (x*y)
print(multi(4,8)) 
```

```python
# Now we gave another example of a lambda function that adds 2
# numbers and return one result. 
add = lambda a , b, c : (a+b+c)
print(add(2,5,9)) 
```

```python
lambda_reciprocal = lambda num: 1 / num     
# using the function defined by lambda keyword  
print( "Lambda keyword: ", lambda_reciprocal(6) ) 
```

```python
# Using Lambda Function with filter()
List_ = [1,5,2,4,6,1,8,2,10,23]
odd_list = list(filter(lambda num : num %2 !=0,List_ ))
even_list = list(filter(lambda num : num %2 ==0,List_ ))
print("Odd Number in List ::",odd_list)
print("Even Number in List ::",even_list)
```

```python
# Using Lambda Function with map()
_list = [2, 4, 5, 1, 3, 7, 8, 9, 10]      
sqrt_list = list(map( lambda num: num ** 2 , _list ))      
print( 'Square of number in list:' ,sqrt_list )    
```

```python
squares = [lambda num = num: num ** 2 for num in range(0, 11)]       
for square in squares:    
       print(square(), end = " ") 
```

```python
Minimum = lambda x, y : x if (x > y) else y       
print('The greater number is:', Minimum( 6,9 ))
```

```python
List  = [ [3, 5, 8, 6], [23, 54, 12, 87], [1, 2, 4, 12, 5] ]        
sort_List = lambda num : ( sorted(n) for n in num )         
third_Largest = lambda num, func : [ l[ len(l) - 2] for l in func(num)]    
result = third_Largest(List, sort_List)      
print('The third largest number from every sub list is:', result )  
```

```python
List = [ [3, 5, 8, 6], [23, 54, 12, 87], [1, 2, 4, 12, 5] ] 
for sublist in List :
    largest_three = sorted(sublist,reverse=True)[:3]
    print(largest_three) 
```
